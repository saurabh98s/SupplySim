```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'S:/SJSU/DATA_226/group_project/daily_time_series_data_with_seasonal_effects.csv'
daily_df = pd.read_csv(file_path)

# Convert the 'Date' column to a datetime object
daily_df['Date'] = pd.to_datetime(daily_df['Date'])
```

```python
import matplotlib.pyplot as plt

# Create a more concise and clearer visualization
plt.figure(figsize=(14, 6))

# Plot daily sales for a subset of SKUs to make the chart more readable
subset_skus = daily_df['SKU'].unique()[:10]  # Select only the first 10 unique SKUs for visualization

# Plot daily sales for the subset of SKUs
for sku in subset_skus:
    product_data = daily_df[daily_df['SKU'] == sku]
    plt.plot(product_data['Date'], product_data['Daily Sales'], label=f"{sku} Sales")

plt.xlabel('Date')
plt.ylabel('Daily Sales')
plt.title('Daily Sales Over Time (Subset of SKUs)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Date' column to a datetime object
daily_df['Date'] = pd.to_datetime(daily_df['Date'])

# Filter data for SKU35
sku35_data = daily_df[daily_df['SKU'] == 'SKU35']

# Plot stock levels for SKU35
plt.figure(figsize=(10, 5))
plt.plot(sku35_data['Date'], sku35_data['Stock Level'], marker='o', label='SKU35 Stock Level')
plt.xlabel('Date')
plt.ylabel('Stock Level')
plt.title('Stock Levels Over Time for SKU35')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
